## Getting Started

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

## Basic Charting

### Reference for Mathematical Expression in Matplotlib
| Command | Result &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; | 
| ----- | ::--------------- 
| `r'$\mathrm{sin}(\frac{x}{2})$'` |  $\mathrm{sin}(\frac{x}{2})$     |
| `r'$\hat y = ax + b$'`       | $\hat y = ax + b$
| `r'$\alpha_i > \beta_i$'` | $\alpha_i > \beta_i$     |
| `r'$\sqrt[3]{x}$'`  | $\sqrt[3]{x}$  |

```python
x = np.linspace(0,10,100)

plt.title(r"$\mathrm{sin}(x) \ vs\  \mathrm{sin}(\frac{x}{2})$")
plt.plot(x, np.sin(x), '--',  color = 'blue', label= "sin(x)")
plt.plot(x, np.sin(x/2), '-', color = '#FFA500', label=r"$\mathrm{sin}(\frac{x}{2})$")
plt.grid(alpha = 0.4, linestyle = '--')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.savefig("figure1.png")
plt.show()
```

## Annotate Your Chart

```python
x = np.linspace(0,10,100)

plt.figure(figsize=(10,10))

plt.title(r"$\mathrm{sin}(x) \ vs\  \mathrm{sin}(\frac{x}{2})$")
plt.plot(x, np.sin(x), color = 'blue', label= "sin(x)")
plt.plot(x, np.sin(x/2), color = '#FFA500', label=r"$\mathrm{sin}(\frac{x}{2})$")
plt.grid(alpha = 0.4, linestyle = '--')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()

plt.annotate('Point 1',xy=(0,0), xytext=(0,-0.25), arrowprops=dict(arrowstyle='->'))
plt.annotate("Point 2", xy=(2/3*math.pi,math.sin(2/3*math.pi)), xytext=(3,0.75), arrowprops=dict(facecolor="yellow",arrowstyle="-|>"),fontsize=16)
plt.annotate("Point 3", xy=(2*math.pi,math.sin(2*math.pi)), xytext=(7,0), arrowprops=dict(facecolor="r",arrowstyle="fancy"),fontsize=12)

plt.savefig("figure2.png")
plt.show()
```

## Highlight Your Chart

```python
x = np.linspace(0,10,100)

plt.title(r"$\mathrm{sin}(x) \ vs\  \mathrm{sin}(\frac{x}{2})$")
plt.plot(x, np.sin(x), color = 'blue', label= "sin(x)")
plt.plot(x, np.sin(x/2), color = '#FFA500', label=r"$\mathrm{sin}(\frac{x}{2})$")
plt.grid(alpha = 0.4, linestyle = '--')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()

plt.axvspan(2, 6, facecolor='red', alpha=0.3)
plt.axhspan(0.75, 1, facecolor='blue', alpha=0.3)

plt.savefig("figure3.png")
plt.show()
```

## Customise xtick in bar chart

```python
labels = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
mobile_sales = [20, 34, 30, 35, 27]
pc_sales = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

plt.bar(x - width/2, mobile_sales, width, label='Mobile')
plt.bar(x + width/2, pc_sales, width, label='PC')

plt.ylabel('Total Sales')
plt.title('Total Sales by Mobile and PC')
plt.xticks([0,1,2,3,4], labels, rotation = "30")

plt.legend()
plt.savefig("figure4.png")
```
